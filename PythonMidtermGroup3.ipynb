```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')
```

```python

print(df.head())


print(df.isnull().sum())


df['release_year'] = df['release_year'].astype(int)
```

**1. Count of Movies vs. TV Shows**

```python
plt.figure(figsize=(8, 5))
df['type'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Count of Movies vs. TV Shows')
plt.ylabel('Number of Titles')
plt.xlabel('Type')
plt.show()
```

**2. Distribution of Release Years**

```python
plt.figure(figsize=(10, 6))
df['release_year'].hist(bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Release Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

**3. Top 10 Countries with Most Titles**

```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='gold')
plt.title('Top 10 Countries with Most Titles')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

**7. Top Directors with Most Titles**

```python
top_directors = df['director'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_directors.plot(kind='bar', color='orange')
plt.title('Top Directors with Most Titles on Netflix')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

**8. Release Year Trends**

**9. Content Type by Country**

**10. Top 10 Actors/Actresses**

```python
actor_counts = df['cast'].str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(10, 6))
actor_counts.plot(kind='bar', color='pink')
plt.title('Top 10 Actors/Actresses on Netflix')
plt.xlabel('Actor/Actress')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

**11. Average Ratings by Genre**

```python

top_genres = df['listed_in'].value_counts().index[:10]
filtered_df = df[df['listed_in'].isin(top_genres)]


main_ratings = ['PG', 'PG-13', 'TV-MA', 'TV-14', 'TV-PG', 'R', 'G']
filtered_df = filtered_df[filtered_df['rating'].isin(main_ratings)]


filtered_df['duration'] = filtered_df['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(14, 7))
sns.boxplot(data=filtered_df, x='listed_in', y='duration', hue='rating')
plt.title('Movie Duration by Top Genres and Rating')
plt.xlabel('Genre')
plt.ylabel('Duration (minutes)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Rating')
plt.show()
```
